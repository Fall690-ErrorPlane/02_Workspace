```python
directory1 = '/content/gdrive/MyDrive/json_image_files/filenames_with_contrails'
for filename in os.listdir(directory1):
  print(filename)
  
  f= open(os.path.join(directory1,filename))
  indiv_json_file = json.loads(f.read()) # reads file in as a python object
  polygon_list = indiv_json_file['polygons'] # list of polygons
  outer_list_counter = 0 # set counter to Zero, this will be the 1st index of the outer list
  inner_list_counter = 0 # set counter to Zero, this will be the 1st index of the inner list
  num_of_contrails_in_scene = len(polygon_list) # count the range of the outer for loop
  num_of_points_along_contrail = len(polygon_list[outer_list_counter]) # count the range for the inner for loop
  current_contrail = polygon_list[outer_list_counter] # list of current contrail
  points_along_contrail_list = polygon_list[outer_list_counter][inner_list_counter] # list of [x,y] coordinates of current contrail

  for contrail in range (len(polygon_list)): # start to loop through the contrails
    print(len(polygon_list))
    print(len(polygon_list[outer_list_counter]))
    x_y_coordinates = np.asarray(polygon_list[outer_list_counter]) # convert to numpy array
    print(x_y_coordinates)
    bbox= BoundingBox(x_y_coordinates)
    print(bbox)
    print(type(bbox)) # this is not serializable
    print(type(bbox.minx)) # this is a numpy array float

    # Must convert bounding box to a serializable python object
    x_1 = bbox.minx
    x_2 = bbox.maxx
    y_1 = bbox.miny
    y_2 = bbox.maxy

    # conversions to regular float
    min_x = x_1.item() 
    max_x = x_2.item()
    min_y = y_1.item()
    max_y = y_2.item()

    #Verify type conversion
    #print(type(min_x)) 
    #print(type(max_x))

    # create list to hold bounding box coordinates
    bounding_box_list = [ min_x,max_x, min_y,max_y] # create list to hold bbox coordinates
    print(bounding_box_list)
    polygon_list[outer_list_counter].append(bounding_box_list)    # add bbox as dictionary
    outer_list_counter = outer_list_counter + 1

  directory2 = '/content/gdrive/MyDrive/json_image_files/json_files_with_bbox_data'
  path = os.path.join(directory2,filename)
  with_bbox_file = open(path, "w")
  indiv_json_file = json.dumps(indiv_json_file, indent= 2 )# converts back to json file
  with_bbox_file.write(indiv_json_file)

#with_bbox_file.close()  


  
```
